```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.pipeline import Pipeline

from sklearn.linear_model import (
    LinearRegression, Ridge, Lasso, ElasticNet, BayesianRidge, HuberRegressor
)
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import (
    RandomForestRegressor, ExtraTreesRegressor,
    GradientBoostingRegressor, AdaBoostRegressor
)
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel as C

from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

import matplotlib.pyplot as plt


CSV_PATH = "hfss_joint_dataset_bandpass.csv"
TARGET_COL = "f0_GHz"
FILTER_OK_ONLY = True


df = pd.read_csv(CSV_PATH)

if FILTER_OK_ONLY and "status" in df.columns:
    df = df[df["status"].astype(str).str.upper() == "OK"].copy()

df = df.dropna(subset=[TARGET_COL]).copy()

feature_cols = ["L1_in", "L2_in", "L3_in", "L4_in"]
missing = [c for c in feature_cols + [TARGET_COL] if c not in df.columns]
if missing:
    raise ValueError(f"CSV missing required columns: {missing}")

X_df = df[feature_cols].astype(float)
y = df[TARGET_COL].astype(float).values

feature_names = X_df.columns.tolist()
X = X_df.values

print("Loaded:", CSV_PATH)
print("Samples:", len(df))
print("Features shape:", X.shape)
print("Target shape  :", y.shape)
print("Feature names :", feature_names)


X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.10,
    random_state=42,
    shuffle=True
)


models = []

models.append((
    "Linear Regression",
    Pipeline([
        ("scaler", StandardScaler()),
        ("reg", LinearRegression())
    ])
))

models.append((
    "Ridge Regression",
    Pipeline([
        ("scaler", StandardScaler()),
        ("reg", Ridge(alpha=1e-4))
    ])
))

models.append((
    "Lasso Regression",
    Pipeline([
        ("scaler", StandardScaler()),
        ("reg", Lasso(alpha=1e-4, max_iter=20000))
    ])
))

models.append((
    "ElasticNet Regression",
    Pipeline([
        ("scaler", StandardScaler()),
        ("reg", ElasticNet(alpha=1e-4, l1_ratio=0.7, max_iter=20000))
    ])
))

models.append((
    "Bayesian Ridge",
    Pipeline([
        ("scaler", StandardScaler()),
        ("reg", BayesianRidge())
    ])
))

models.append((
    "Huber Regressor",
    Pipeline([
        ("scaler", StandardScaler()),
        ("reg", HuberRegressor(max_iter=5000, epsilon=1.2))
    ])
))

models.append((
    "Polynomial (deg 3)",
    Pipeline([
        ("poly", PolynomialFeatures(degree=3, include_bias=False)),
        ("scaler", StandardScaler(with_mean=False)),
        ("reg", LinearRegression())
    ])
))

models.append((
    "kNN Regressor (k=7)",
    Pipeline([
        ("scaler", StandardScaler()),
        ("reg", KNeighborsRegressor(n_neighbors=7, weights="distance"))
    ])
))

models.append((
    "Decision Tree",
    DecisionTreeRegressor(
        random_state=42,
        min_samples_split=4,
        min_samples_leaf=2
    )
))

models.append((
    "Random Forest",
    RandomForestRegressor(
        n_estimators=800,
        max_depth=None,
        min_samples_split=4,
        min_samples_leaf=2,
        n_jobs=-1,
        random_state=42
    )
))

models.append((
    "Extra Trees",
    ExtraTreesRegressor(
        n_estimators=800,
        max_depth=None,
        min_samples_split=4,
        min_samples_leaf=2,
        n_jobs=-1,
        random_state=42
    )
))

models.append((
    "Gradient Boosting",
    GradientBoostingRegressor(
        n_estimators=800,
        learning_rate=0.03,
        max_depth=3,
        subsample=0.9,
        random_state=42
    )
))

models.append((
    "AdaBoost Regressor",
    AdaBoostRegressor(
        estimator=DecisionTreeRegressor(max_depth=3),
        n_estimators=900,
        learning_rate=0.03,
        random_state=42
    )
))

models.append((
    "SVR (RBF)",
    Pipeline([
        ("scaler", StandardScaler()),
        ("reg", SVR(kernel="rbf", C=100.0, gamma="scale"))
    ])
))

models.append((
    "Neural Network (MLP)",
    Pipeline([
        ("scaler", StandardScaler()),
        ("reg", MLPRegressor(
            hidden_layer_sizes=(256, 256, 128),
            activation="relu",
            solver="adam",
            alpha=1e-4,
            learning_rate_init=1e-3,
            max_iter=12000,
            batch_size=32,
            random_state=42
        ))
    ])
))

kernel = C(1.0, (1e-3, 1e3)) * RBF(length_scale=1.0)
models.append((
    "Gaussian Process",
    Pipeline([
        ("scaler", StandardScaler()),
        ("reg", GaussianProcessRegressor(
            kernel=kernel,
            alpha=1e-6,
            normalize_y=True,
            random_state=42
        ))
    ])
))


print("\n===== 90/10 Train-Test Model Comparison (target: {}) =====".format(TARGET_COL))

metrics_list = []
predictions_store = {}
fitted_models = {}

for name, model in models:
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)

    metrics_list.append({"name": name, "r2": r2, "mae": mae, "mse": mse, "rmse": rmse})
    predictions_store[name] = y_pred
    fitted_models[name] = model

    print(f"\n{name}")
    print(f"Test R²   = {r2: .4f}")
    print(f"MAE       = {mae: .4f}")
    print(f"MSE       = {mse: .4f}")
    print(f"RMSE      = {rmse: .4f}")


model_names = [m["name"] for m in metrics_list]
r2_values = [m["r2"] for m in metrics_list]

plt.figure(figsize=(10, 5))
plt.bar(range(len(model_names)), r2_values)
plt.xticks(range(len(model_names)), model_names, rotation=60, ha="right")
plt.ylabel("Test R²")
plt.title(f"Model Comparison (target: {TARGET_COL})")
plt.tight_layout()
plt.show()

best_entry = max(metrics_list, key=lambda d: d["r2"])
best_name = best_entry["name"]
best_pred = predictions_store[best_name]
best_model = fitted_models[best_name]

print(f"\nBest model by R²: {best_name} (R² = {best_entry['r2']:.4f})")

plt.figure(figsize=(6, 6))
plt.scatter(y_test, best_pred, alpha=0.7)
min_val = min(float(np.min(y_test)), float(np.min(best_pred)))
max_val = max(float(np.max(y_test)), float(np.max(best_pred)))
plt.plot([min_val, max_val], [min_val, max_val], linestyle="--")
plt.xlabel("True values")
plt.ylabel("Predicted values")
plt.title(f"{best_name} – True vs Predicted ({TARGET_COL})")
plt.tight_layout()
plt.show()

sorted_idx = np.argsort(y_test)
y_true_sorted = y_test[sorted_idx]
y_pred_sorted = best_pred[sorted_idx]

plt.figure(figsize=(8, 4))
plt.plot(y_true_sorted, label="True", linewidth=2)
plt.plot(y_pred_sorted, label="Predicted", linestyle="--")
plt.xlabel("Sample index (sorted by true value)")
plt.ylabel(TARGET_COL)
plt.title(f"{best_name} – Sorted True vs Predicted")
plt.legend()
plt.tight_layout()
plt.show()

residuals = y_test - best_pred

plt.figure(figsize=(7, 4))
plt.hist(residuals, bins=20, edgecolor="black", alpha=0.8)
plt.xlabel("Residual (True - Predicted)")
plt.ylabel("Count")
plt.title(f"{best_name} – Residual Histogram")
plt.tight_layout()
plt.show()

plt.figure(figsize=(7, 4))
plt.scatter(best_pred, residuals, alpha=0.7)
plt.axhline(0.0, linestyle="--")
plt.xlabel("Predicted values")
plt.ylabel("Residual (True - Predicted)")
plt.title(f"{best_name} – Residuals vs Predicted")
plt.tight_layout()
plt.show()

top_models_sorted = sorted(metrics_list, key=lambda d: d["r2"], reverse=True)
top_models_names = [m["name"] for m in top_models_sorted[:5]]

abs_errors_list = []
for name in top_models_names:
    y_pred_model = predictions_store[name]
    abs_errors_list.append(np.abs(y_test - y_pred_model))

plt.figure(figsize=(8, 4))
plt.boxplot(abs_errors_list, labels=top_models_names, showfliers=False)
plt.ylabel("|Error|")
plt.title(f"Absolute Error Distribution – Top Models (target: {TARGET_COL})")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()

def get_tree_regressor_from_pipeline(model):
    if isinstance(model, Pipeline):
        return model.named_steps.get("reg", None)
    return model

tree_reg = get_tree_regressor_from_pipeline(best_model)

if hasattr(tree_reg, "feature_importances_"):
    importances = tree_reg.feature_importances_
    if len(importances) == len(feature_names):
        plt.figure(figsize=(8, 4))
        idx_sorted = np.argsort(importances)[::-1]
        sorted_names = [feature_names[i] for i in idx_sorted]
        sorted_importances = importances[idx_sorted]
        plt.bar(range(len(sorted_importances)), sorted_importances)
        plt.xticks(range(len(sorted_importances)), sorted_names, rotation=0)
        plt.ylabel("Importance")
        plt.title(f"{best_name} – Feature Importances")
        plt.tight_layout()
        plt.show()
    else:
        print("\n[Info] Feature importances length mismatch, skipping importance plot.")
else:
    print(f"\n[Info] Best model ({best_name}) has no feature_importances_; skipping feature importance plot.")
```

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.pipeline import Pipeline

from sklearn.linear_model import (
    LinearRegression, Ridge, Lasso, ElasticNet, BayesianRidge, HuberRegressor
)
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import (
    RandomForestRegressor, ExtraTreesRegressor,
    GradientBoostingRegressor, AdaBoostRegressor
)
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel as C

from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt


CSV_PATHS = [
    "hfss_joint_dataset_bandpass.csv",
    "hfss_joint_dataset_bandpass_2.csv",
]

TARGET_COL = "f0_GHz"
FILTER_OK_ONLY = True

MLP_EPOCHS = 50000
MLP_EARLY_STOPPING = True
MLP_NO_IMPROVE = 30

TEST_SIZE = 0.10
EXTREME_MODE = "quantile"
EXTREME_Q = 0.02
BOUND_MARGIN = 0.0


dfs = []
for p in CSV_PATHS:
    d = pd.read_csv(p)
    d["__source_csv__"] = p
    dfs.append(d)

df = pd.concat(dfs, ignore_index=True)
df = df.drop_duplicates().reset_index(drop=True)

if FILTER_OK_ONLY and "status" in df.columns:
    df = df[df["status"].astype(str).str.upper() == "OK"].copy()

df = df.dropna(subset=[TARGET_COL]).copy()

feature_cols = ["L1_in", "L2_in", "L3_in", "L4_in"]
missing = [c for c in feature_cols + [TARGET_COL] if c not in df.columns]
if missing:
    raise ValueError(f"Combined CSVs missing required columns: {missing}")

X_df = df[feature_cols].astype(float)
y = df[TARGET_COL].astype(float).values

feature_names = X_df.columns.tolist()
X = X_df.values

print("Loaded & combined:", CSV_PATHS)
print("Samples:", len(df))
print("Features shape:", X.shape)
print("Target shape  :", y.shape)
print("Feature names :", feature_names)
print(df["__source_csv__"].value_counts() if "__source_csv__" in df.columns else "")


rng = np.random.default_rng(42)

if EXTREME_MODE == "quantile":
    q_low = X_df.quantile(EXTREME_Q)
    q_high = X_df.quantile(1.0 - EXTREME_Q)
    extreme_mask = (X_df.lt(q_low) | X_df.gt(q_high)).any(axis=1).to_numpy()
elif EXTREME_MODE == "bounds":
    mins = X_df.min()
    maxs = X_df.max()
    extreme_mask = ((X_df.le(mins + BOUND_MARGIN)) | (X_df.ge(maxs - BOUND_MARGIN))).any(axis=1).to_numpy()
else:
    raise ValueError("EXTREME_MODE must be 'quantile' or 'bounds'")

idx_all = np.arange(len(df))
idx_ext = idx_all[extreme_mask]
idx_in = idx_all[~extreme_mask]

n_total = len(idx_all)
n_test_target = int(np.ceil(TEST_SIZE * n_total))

idx_test = idx_ext.copy()

if len(idx_test) < n_test_target:
    n_need = n_test_target - len(idx_test)
    n_need = min(n_need, len(idx_in))
    idx_test_extra = rng.choice(idx_in, size=n_need, replace=False)
    idx_test = np.concatenate([idx_test, idx_test_extra])

idx_test_set = set(idx_test.tolist())
idx_train = np.array([i for i in idx_in if i not in idx_test_set], dtype=int)

X_train, y_train = X[idx_train], y[idx_train]
X_test, y_test = X[idx_test], y[idx_test]

print("\nSplit summary (extremities excluded from training):")
print(f"  Total samples        : {n_total}")
print(f"  Extremities (held out): {len(idx_ext)}")
print(f"  Train samples        : {len(idx_train)}")
print(f"  Test samples         : {len(idx_test)}")
print(f"  Effective test frac  : {len(idx_test)/n_total:.3f}")


models = []

models.append((
    "Linear Regression",
    Pipeline([("scaler", StandardScaler()), ("reg", LinearRegression())])
))

models.append((
    "Ridge Regression",
    Pipeline([("scaler", StandardScaler()), ("reg", Ridge(alpha=1e-4))])
))

models.append((
    "Lasso Regression",
    Pipeline([("scaler", StandardScaler()), ("reg", Lasso(alpha=1e-4, max_iter=20000))])
))

models.append((
    "ElasticNet Regression",
    Pipeline([("scaler", StandardScaler()), ("reg", ElasticNet(alpha=1e-4, l1_ratio=0.7, max_iter=20000))])
))

models.append((
    "Bayesian Ridge",
    Pipeline([("scaler", StandardScaler()), ("reg", BayesianRidge())])
))

models.append((
    "Huber Regressor",
    Pipeline([("scaler", StandardScaler()), ("reg", HuberRegressor(max_iter=5000, epsilon=1.2))])
))

models.append((
    "Polynomial (deg 3)",
    Pipeline([("poly", PolynomialFeatures(degree=3, include_bias=False)),
              ("scaler", StandardScaler(with_mean=False)),
              ("reg", LinearRegression())])
))

models.append((
    "kNN Regressor (k=7)",
    Pipeline([("scaler", StandardScaler()), ("reg", KNeighborsRegressor(n_neighbors=7, weights="distance"))])
))

models.append((
    "Decision Tree",
    DecisionTreeRegressor(random_state=42, min_samples_split=4, min_samples_leaf=2)
))

models.append((
    "Random Forest",
    RandomForestRegressor(n_estimators=800, max_depth=None, min_samples_split=4, min_samples_leaf=2, n_jobs=-1, random_state=42)
))

models.append((
    "Extra Trees",
    ExtraTreesRegressor(n_estimators=800, max_depth=None, min_samples_split=4, min_samples_leaf=2, n_jobs=-1, random_state=42)
))

models.append((
    "Gradient Boosting",
    GradientBoostingRegressor(n_estimators=800, learning_rate=0.03, max_depth=3, subsample=0.9, random_state=42)
))

models.append((
    "AdaBoost Regressor",
    AdaBoostRegressor(estimator=DecisionTreeRegressor(max_depth=3), n_estimators=900, learning_rate=0.03, random_state=42)
))

models.append((
    "SVR (RBF)",
    Pipeline([("scaler", StandardScaler()), ("reg", SVR(kernel="rbf", C=100.0, gamma="scale"))])
))

models.append((
    "Neural Network (MLP)",
    Pipeline([("scaler", StandardScaler()),
              ("reg", MLPRegressor(hidden_layer_sizes=(256, 256, 128),
                                   activation="relu",
                                   solver="adam",
                                   alpha=1e-4,
                                   learning_rate_init=1e-3,
                                   max_iter=MLP_EPOCHS,
                                   batch_size=32,
                                   early_stopping=MLP_EARLY_STOPPING,
                                   n_iter_no_change=MLP_NO_IMPROVE,
                                   validation_fraction=0.15,
                                   random_state=42,
                                   verbose=False))])
))

kernel = C(1.0, (1e-3, 1e3)) * RBF(length_scale=1.0)
models.append((
    "Gaussian Process",
    Pipeline([("scaler", StandardScaler()),
              ("reg", GaussianProcessRegressor(kernel=kernel, alpha=1e-6, normalize_y=True, random_state=42))])
))


print("\n===== Train-Test Model Comparison (target: {}) =====".format(TARGET_COL))

metrics_list = []
predictions_store = {}
fitted_models = {}

for name, model in models:
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)

    metrics_list.append({"name": name, "r2": r2, "mae": mae, "mse": mse, "rmse": rmse})
    predictions_store[name] = y_pred
    fitted_models[name] = model

    print(f"\n{name}")
    print(f"Test R²   = {r2: .4f}")
    print(f"MAE       = {mae: .4f}")
    print(f"MSE       = {mse: .4f}")
    print(f"RMSE      = {rmse: .4f}")


model_names = [m["name"] for m in metrics_list]
r2_values = [m["r2"] for m in metrics_list]

plt.figure(figsize=(10, 5))
plt.bar(range(len(model_names)), r2_values)
plt.xticks(range(len(model_names)), model_names, rotation=60, ha="right")
plt.ylabel("Test R²")
plt.title(f"Model Comparison (target: {TARGET_COL})")
plt.tight_layout()
plt.show()

best_entry = max(metrics_list, key=lambda d: d["r2"])
best_name = best_entry["name"]
best_pred = predictions_store[best_name]
best_model = fitted_models[best_name]

print(f"\nBest model by R²: {best_name} (R² = {best_entry['r2']:.4f})")

plt.figure(figsize=(6, 6))
plt.scatter(y_test, best_pred, alpha=0.7)
min_val = min(float(np.min(y_test)), float(np.min(best_pred)))
max_val = max(float(np.max(y_test)), float(np.max(best_pred)))
plt.plot([min_val, max_val], [min_val, max_val], linestyle="--")
plt.xlabel("True values")
plt.ylabel("Predicted values")
plt.title(f"{best_name} – True vs Predicted ({TARGET_COL})")
plt.tight_layout()
plt.show()

sorted_idx = np.argsort(y_test)
y_true_sorted = y_test[sorted_idx]
y_pred_sorted = best_pred[sorted_idx]

plt.figure(figsize=(8, 4))
plt.plot(y_true_sorted, label="True", linewidth=2)
plt.plot(y_pred_sorted, label="Predicted", linestyle="--")
plt.xlabel("Sample index (sorted by true value)")
plt.ylabel(TARGET_COL)
plt.title(f"{best_name} – Sorted True vs Predicted")
plt.legend()
plt.tight_layout()
plt.show()

residuals = y_test - best_pred

plt.figure(figsize=(7, 4))
plt.hist(residuals, bins=20, edgecolor="black", alpha=0.8)
plt.xlabel("Residual (True - Predicted)")
plt.ylabel("Count")
plt.title(f"{best_name} – Residual Histogram")
plt.tight_layout()
plt.show()

plt.figure(figsize=(7, 4))
plt.scatter(best_pred, residuals, alpha=0.7)
plt.axhline(0.0, linestyle="--")
plt.xlabel("Predicted values")
plt.ylabel("Residual (True - Predicted)")
plt.title(f"{best_name} – Residuals vs Predicted")
plt.tight_layout()
plt.show()

top_models_sorted = sorted(metrics_list, key=lambda d: d["r2"], reverse=True)
top_models_names = [m["name"] for m in top_models_sorted[:5]]

abs_errors_list = []
for name in top_models_names:
    y_pred_model = predictions_store[name]
    abs_errors_list.append(np.abs(y_test - y_pred_model))

plt.figure(figsize=(8, 4))
plt.boxplot(abs_errors_list, labels=top_models_names, showfliers=False)
plt.ylabel("|Error|")
plt.title(f"Absolute Error Distribution – Top Models (target: {TARGET_COL})")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()

def get_tree_regressor_from_pipeline(model):
    if isinstance(model, Pipeline):
        return model.named_steps.get("reg", None)
    return model

tree_reg = get_tree_regressor_from_pipeline(best_model)

if hasattr(tree_reg, "feature_importances_"):
    importances = tree_reg.feature_importances_
    if len(importances) == len(feature_names):
        plt.figure(figsize=(8, 4))
        idx_sorted = np.argsort(importances)[::-1]
        sorted_names = [feature_names[i] for i in idx_sorted]
        sorted_importances = importances[idx_sorted]
        plt.bar(range(len(sorted_importances)), sorted_importances)
        plt.xticks(range(len(sorted_importances)), sorted_names, rotation=0)
        plt.ylabel("Importance")
        plt.title(f"{best_name} – Feature Importances")
        plt.tight_layout()
        plt.show()
    else:
        print("\n[Info] Feature importances length mismatch, skipping importance plot.")
else:
    print(f"\n[Info] Best model ({best_name}) has no feature_importances_; skipping feature importance plot.")
```
